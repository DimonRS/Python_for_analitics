#### _**3.1 Найдите среднюю стоимость домов в зависимости от количества спален.  Отсортируйте от меньшей стоимости к большей**_ 

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

#### _**3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома**_ 

```python
condition = df[['price', 'condition']].groupby('condition')
condition.agg({'price': ['min', 'mean', 'max']})
```

```python
df.pivot_table(index = 'condition', values = 'price', aggfunc = ['min', 'mean', 'max'])
```
