### Задача 2:

_**На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе, при помощи использования kwargs**_

```python
def total(**kwargs):
    total = sum (kwargs.values())
    return total

print('Общее количество фруктов на складе =', total (яблоки = 7, бананы = 3, манго = 6, мандарины = 9, персики = 2, хурма = 8))
```

### Задача 3:
_**Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты**_

[100, 125, -90, 345, 655, -1, 0, 200]

_**Используйте list comprehensions**_

```python
adv_cost = [100, 125, -90, 345, 655, -1, 0, 200]
```

```python
correct_adv_cost = [i for i in adv_cost if i >= 0]
print ('Список без отрицательных сумм: ', correct_adv_cost)
print('Суммарные затраты =', sum(correct_adv_cost))
```

### Найти 4.2   
_**Найдите выручку компании в зависимости от месяца.**_  
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.  
Используйте аннотирование типов.

```python
def transform_months (d: list, r: list) -> dict:
    result = {}
    m_name = ['январь', 'февраль', 'март', 'апрель', 'май', 'июнь', 'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь']
    m_txt = ['-01-', '-02-', '-03-', '-04-', '-05-', '-06-', '-07-', '-08-', '-09-', '-10-', '-11-', '-12-']
    for m in range(12):
        r_s = sum(revenue[i] for i in range(len(r)) if m_txt[m] in d[i])
        result.update({m_name[m] : r_s })
    return(result)

print ('Выручка по месяцам:\n', transform_months (date, revenue))
```
